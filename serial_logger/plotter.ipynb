```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f=open('data_left.txt','r')
```

```python
rpm_list=[]
pwm_list=[]
while True:
    data=f.readline()
    if data == '':
        break
    data_list=data.split(',')
    rpm=float(data_list[0].split(':')[-1])
    pwm=float(data_list[1].split(':')[-1])
    # print(f'{rpm} {pwm}')
    rpm_list.append(rpm)
    pwm_list.append(pwm)

```

```python
plt.figure(dpi=200)
plt.plot(pwm_list,rpm_list)
plt.xlabel('pwm')
plt.ylabel('rpm')
```

```python
plt.figure(dpi=200)
coef = np.polyfit(pwm_list[50:],rpm_list[50:],1)
poly1d_fn = np.poly1d(coef) 
plt.plot(pwm_list,rpm_list, 'r.', pwm_list, poly1d_fn(pwm_list), '--k',markersize=1)
plt.xlabel('pwm')
plt.ylabel('rpm')
plt.legend(['real data','fit data'])
```
